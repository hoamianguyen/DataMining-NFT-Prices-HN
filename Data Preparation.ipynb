```python
df['Datetime_updated_minutes'] = df['Datetime_updated_seconds'].values.astype('<M8[m]')
```

```python
ETHUSD['Datetime_updated_minutes'] = pd.to_datetime(ETHUSD['Date'],dayfirst=True)
```

```python
ETHUSD.info()
```

```python
df
```

```python
MergedDataset=df.merge(ETHUSD, how='left', on='Datetime_updated_minutes')
```

```python
MergedDataset=MergedDataset.drop(['Date','Symbol','Unique_id_collection','Datetime_updated','Datetime_updated_seconds'],
                                 axis=1)
```

```python
MergedDataset
```

```python
MergedDataset.to_csv('MergedDataset.csv')
```

```python
ColCatDate = MergedDataset.groupby(['Collection_cleaned', 'Category','Datetime_updated_minutes'],as_index=False).sum()
```

```python
ColCatDate = ColCatDate.drop(['ID_token','Price_Crypto','Open','High','Low','Close','Volume'],
                             axis=1)
```

```python
ColCatDate=ColCatDate.rename(columns={'Price_USD':'Transaction Volume'})
```

```python
MergedDatasetVol=MergedDataset.merge(ColCatDate, how='left',
                                     on=['Collection_cleaned','Category','Datetime_updated_minutes'])
```

```python
MergedDatasetVol
```

```python
ColCatDateM = MergedDataset.groupby(['Collection_cleaned', 'Category','Datetime_updated_minutes'],as_index=False).mean()
```

```python
ColCatDateM = ColCatDateM.drop(['ID_token','Price_Crypto','Open','High','Low','Close','Volume'],
                             axis=1)
```

```python
ColCatDateM=ColCatDateM.rename(columns={'Price_USD':'Average Price'})
```

```python
MergedDatasetVolAvg=MergedDatasetVol.merge(ColCatDateM, how='left',
                                     on=['Collection_cleaned','Category','Datetime_updated_minutes'])
```

```python
MergedDatasetVolAvg
```

```python
MergedDatasetVolAvg['Log Variance'] = np.log(MergedDatasetVolAvg['High']/MergedDatasetVolAvg['Low'])
```

```python
PrepDataset=MergedDatasetVolAvg
```

```python
PrepDataset.to_csv('PrepDataset.csv')
```
